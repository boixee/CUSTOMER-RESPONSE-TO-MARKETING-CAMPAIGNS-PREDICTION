# Inconsistent Data

```python
cr['Date_Customer'] = pd.to_datetime(cr['Date_Customer'])
```

```python
cr['Education_Level'].value_counts()
```

```python
cr['Education_Level'] = cr['Education_Level'].str.replace("Graduation", "Degree")
cr['Education_Level'] = cr['Education_Level'].str.replace("2n Cycle", "Secondary")
cr['Education_Level'] = cr['Education_Level'].str.replace("Basic", "Primary")
```

```python
cr
```

```python
cr['Marital_Status'].value_counts()
```

```python
cr['Marital_Status'] = cr['Marital_Status'].str.replace('Together', 'Married')
cr['Marital_Status'] = cr['Marital_Status'].str.replace('Alone', 'Single')
cr['Marital_Status'] = cr['Marital_Status'].str.replace('Divorced', 'Widow')
cr['Marital_Status'] = cr['Marital_Status'].replace(['YOLO', 'Absurd'], 'Unspecified')
```

# Dealing with Categorical Inputs

```python
cr['Marital_Status'] = cr['Marital_Status'].astype('category')
cr['Education_Level'] = cr['Education_Level'].astype('category')
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
cr['Marital_Status'] = encoder.fit_transform(cr['Marital_Status'])
cr['Education_Level'] = encoder.fit_transform(cr['Education_Level'])
```

```python
cr.head()
```

# Feature Engineering

```python
cr['No_of_child_in_home'] = cr['No_of_Kids_in_home'] + cr['No_of_Teen_in_home']
```

```python
cr['Age_of_Customer'] = 2023 - cr['Year_of_Birth']
```

```python
today = pd.to_datetime('2023-09-07')
cr['duration_in_days'] = (today - cr['Date_Customer']).dt.days
```

```python
cr['Year'] = cr['Date_Customer'].dt.year
cr['Month'] = cr['Date_Customer'].dt.month
cr['Day'] = cr['Date_Customer'].dt.day
```

```python
cr.head()
```

```python
cr = cr.drop(['No_of_Kids_in_home','No_of_Teen_in_home','Year_of_Birth','Date_Customer'], axis=1)
cr.head()
```
